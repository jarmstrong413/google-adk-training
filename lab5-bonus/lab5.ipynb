```python
import json
import vertexai
from google.oauth2 import service_account
from google.colab import userdata

vertexai.init(
    project="qwiklabs-gcp-00-819951dd6b01",
    location="us-central1",
    staging_bucket="gs://adk-capstone"
)


from google.adk.agents import Agent
from google.adk.tools import google_search

search_agent = Agent(
    name="search_bot",
    model="gemini-2.5-flash",
    description=("Agent that lets users search Google"),
    instruction=("""You are a helpful chatbot. Return relevant, well-structured information that you retrieve from google_search."""),
    tools=[google_search]
)
```

```python
from vertexai.preview.reasoning_engines import AdkApp

app = AdkApp(agent=search_agent)

for event in app.stream_query(
    user_id="user001",
    message="Give me highlights from the winter olympics 2026",
):
  print(event)
```

```python
from vertexai import agent_engines
remote_agent = agent_engines.create(
    app,
    requirements=["google-cloud-aiplatform[agent_engines,adk]"]
)
```

```python
for event in remote_agent.stream_query(
    user_id="agent_engine_test",
    message="Give me highlights from the winter olympics 2026",
):
  print(event)
```
